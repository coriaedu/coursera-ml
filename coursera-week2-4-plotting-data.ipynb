## Plotting Data

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, .98, 99)
t
```

```python
y1 = np.sin(2*np.pi*4*t)
```

```python
plt.plot(t,y1)
plt.show()
```

```python
y2 = np.cos(2*np.pi*4*t)
plt.plot(t,y2, 'r')
plt.show()
```

```python
plt.plot(t,y1)
plt.plot(t,y2, 'r')
plt.show()
```

```python
plt.plot(t,y1)
plt.plot(t,y2, 'r')
plt.xlabel('time')
plt.ylabel('value')
plt.legend(['sin', 'cos'])
plt.title('my plot')
p = plt;
p.show()
```

```python
plt.figure(1)
plt.plot(t,y1)
plt.figure(2)
plt.plot(t,y2, 'r')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1)
plt.subplot(1,2,2)
plt.plot(t,y2, 'r')
plt.axis([0.5, 1, -.5, .5])
plt.show()
```
